```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
exact_ice_P_kv = []
pred_ice_P_kv = []
exact_ice_P_old = []
pred_ice_P_old = []
with open('../Data/ice_data/test_kv.dat', 'r') as file:
    for line in file:
        exact_ice_P_values = -float(line.split()[4])
        exact_ice_P_kv.append(exact_ice_P_values)

        pred_ice_P_values = -float(line.split()[5])
        pred_ice_P_kv.append(pred_ice_P_values)

with open('../Data/ice_data/test_old.dat', 'r') as file:
    for line in file:
        exact_ice_P_values = -float(line.split()[4])
        exact_ice_P_old.append(exact_ice_P_values)

        pred_ice_P_values = -float(line.split()[5])
        pred_ice_P_old.append(pred_ice_P_values)

print(len(exact_ice_P_kv))
print(len(exact_ice_P_old))
```

```python
exact_LGPS_P_kv = []
pred_LGPS_P_kv = []
exact_LGPS_P_old = []
pred_LGPS_P_old = []
with open('../Data/LGPS_data/test_kv.dat', 'r') as file:
    for line in file:
        exact_LGPS_P_values = -float(line.split()[4])
        exact_LGPS_P_kv.append(exact_LGPS_P_values)

        pred_LGPS_P_values = -float(line.split()[5])
        pred_LGPS_P_kv.append(pred_LGPS_P_values)

with open('../Data/LGPS_data/test_old.dat', 'r') as file:
    for line in file:
        exact_LGPS_P_values = -float(line.split()[4])
        exact_LGPS_P_old.append(exact_LGPS_P_values)

        pred_LGPS_P_values = -float(line.split()[5])
        pred_LGPS_P_old.append(pred_LGPS_P_values)

print(len(exact_LGPS_P_kv))
print(len(exact_LGPS_P_old))
```

```python
exact_lqBN_P_kv = []
pred_lqBN_P_kv = []
exact_lqBN_P_old = []
pred_lqBN_P_old = []
with open('../Data/lqBN_data/test_kv.dat', 'r') as file:
    for line in file:
        exact_lqBN_P_values = -float(line.split()[4])
        exact_lqBN_P_kv.append(exact_lqBN_P_values)

        pred_lqBN_P_values = -float(line.split()[5])
        pred_lqBN_P_kv.append(pred_lqBN_P_values)

with open('../Data/lqBN_data/test_old.dat', 'r') as file:
    for line in file:
        exact_lqBN_P_values = -float(line.split()[4])
        exact_lqBN_P_old.append(exact_lqBN_P_values)

        pred_lqBN_P_values = -float(line.split()[5])
        pred_lqBN_P_old.append(pred_lqBN_P_values)

print(len(exact_lqBN_P_kv))
print(len(exact_lqBN_P_old))
```

```python
f, axes = plt.subplots(3, 1, dpi=100)

f.set_size_inches((7, 15))
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)

f.tight_layout()
f.text(-0.1, 0.975, 'a', fontsize=20, fontweight='bold')
f.text(-0.1, 0.645, 'b', fontsize=20, fontweight='bold')
f.text(-0.1, 0.315, 'c', fontsize=20, fontweight='bold')

## liquid BN ##
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)
axes[0].spines['bottom'].set_linewidth(1.5)
axes[0].spines['left'].set_linewidth(1.5)
axes[0].tick_params(width=1.5)
axes[0].tick_params(axis='x', labelsize=18)
axes[0].tick_params(axis='y', labelsize=18)
axes[0].scatter(exact_lqBN_P_old[:], pred_lqBN_P_old[:], color='C3', edgecolor='white', label='Scheme1')
axes[0].scatter(exact_lqBN_P_kv[:], pred_lqBN_P_kv[:], color='C0', edgecolor='white', label='Scheme2')
axes[0].plot(np.arange(0, 45, 0.1), np.arange(0, 45, 0.1), color='black', linestyle='-')

## LGPS ##
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].spines['bottom'].set_linewidth(1.5)
axes[1].spines['left'].set_linewidth(1.5)
axes[1].tick_params(width=1.5)
axes[1].tick_params(axis='x', labelsize=18)
axes[1].tick_params(axis='y', labelsize=18)
axes[1].scatter(exact_LGPS_P_old[:], pred_LGPS_P_old[:], color='C3', edgecolor='white', label='Scheme1')
axes[1].scatter(exact_LGPS_P_kv[:], pred_LGPS_P_kv[:], color='C0', edgecolor='white', label='Scheme2')
axes[1].plot(np.arange(-2., 2.5, 0.1), np.arange(-2., 2.5, 0.1), color='black', linestyle='-', label='y=x line')
axes[1].set_ylabel('Predicted Pressure (GPa)', fontsize=20)

## ice ##
axes[2].spines['top'].set_visible(False)
axes[2].spines['right'].set_visible(False)
axes[2].spines['bottom'].set_linewidth(1.5)
axes[2].spines['left'].set_linewidth(1.5)
axes[2].tick_params(width=1.5)
axes[2].tick_params(axis='x', labelsize=18)
axes[2].tick_params(axis='y', labelsize=18)
axes[2].scatter(exact_ice_P_old[:], pred_ice_P_old[:], color='C3', edgecolor='white', label='Scheme1')
axes[2].scatter(exact_ice_P_kv[:], pred_ice_P_kv[:], color='C0', edgecolor='white', label='Scheme2')
axes[2].plot(np.arange(0.2, 1., 0.1), np.arange(0.2, 1., 0.1), color='black', linestyle='-')
axes[2].set_xlabel('Reference Pressure (GPa)', fontsize=20)

plt.show()
```
